```python
book_data=book_data.drop(book_data.columns[[0]],axis=1)
book_data
```

```python
book_data=book_data.rename(columns={"User.ID":"userid","Book.Title":"booktitle","Book.Rating":"bookrating"})
```

```python
#number of unique users in the dataset
len(book_data.userid.unique())
```

```python
#number of unique books in the dataset
len(book_data.booktitle.unique())
```

```python
user_books_df = book_data.pivot_table(index='userid',
                                 columns='booktitle',
                                 values='bookrating').reset_index(drop=True)
user_books_df
```

```python
user_books_df.shape
```

```python
user_books_df
```

```python
#replce those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_books_df.values,metric='cosine')
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[10:40]
```
