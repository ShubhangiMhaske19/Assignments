**Kmeans**


```python
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("crime_data.csv")
data
```

```python
# Similar to the sklearn.preprocessing, MinMaxScaler function
def minmax_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Similar to the sklearn.preprocessing, StandardScaler function
def norm_func(i):
    x = (i-i.mean())/(i.std())
    return (x)
```

```python
x=norm_func(data.iloc[:,1:])
x
```

```python
#WCSS (Within Cluster Sum of Squares)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

print(wcss)    
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=0)
clusters_new.fit(x)
```

```python
#Assign clusters to the data set
data['clusterid_new'] = clusters_new.labels_
data[0:5]
```

```python
data.groupby('clusterid_new').mean()
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
data = pd.read_csv("crime_data.csv")
data[0:5]
```

```python
data1=data.iloc[:,1:]
data1
```

```python
from sklearn.preprocessing import StandardScaler

stscaler = StandardScaler()
X=stscaler.fit_transform(data1.values)
X
```

```python
dbscan = DBSCAN(eps=2, min_samples=8)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
data['Cluster']=pd.DataFrame(dbscan.labels_)
data
```

```python
data['Cluster'].value_counts()
```

```python
import matplotlib.pyplot as plt

def dbscan(X,eps,min_samples):
    ss=StandardScaler()
    X=ss.fit_transform(X)
    db=DBSCAN(eps=eps,min_samples=min_samples)
    db.fit(X)
    y_pred=db.fit_predict(X)
    plt.scatter(X[:,0],X[:,1],c=y_pred, cmap='Paired')
    plt.title("DBSCAN")
```

```python
dbscan(data1.values,1.0,8)
```
